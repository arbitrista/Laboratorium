# Servitium Publicum

>  El **servicio público** es la actividad o conjunto de prestaciones que el Estado o entidades gubernamentales proporcionan a los ciudadanos para satisfacer necesidades colectivas, garantizando el bienestar y el interés general, bajo principios de accesibilidad, continuidad y equidad.

## Evolución Nómina Pública

> La **Nómina Pública** es el conjunto de registros que detalla los salarios, beneficios y deducciones de los empleados que trabajan para el Estado o entidades gubernamentales, gestionando su remuneración de acuerdo con las normativas y políticas del sector público.

```python
tables = camelot.read_pdf("/home/dbremont/Downloads/Empleados-Fijos-1.pdf", pages='all', flavor='stream')
```

```python
# Create a list to hold DataFrames
dataframes = []

# Loop through the extracted tables and append DataFrames
for i in range(len(tables)):
    df = tables[i].df  # Get the DataFrame from the Camelot Table
    dataframes.append(df)
```

```python
combined_df = pd.concat(dataframes, ignore_index=True)
column_index = 3
column_name = combined_df.columns[column_index]
cleaned_df = combined_df[~(combined_df[column_name].isna() | (combined_df[column_name] == '') | (combined_df[column_name].str.len() <= 2))]
```
